### Relationship
There are many diffrerent chart types that you can use to understand relationships between variables in your data.
<ul>
    <li>sns.barplot: <strong>Bar charts</strong> are useful for comparubg quantities corresponding to different groups</li>
    <li>sns.heatmap: <strong>Heatmaps</strong> can be used to find color-coded patterns in tables of numbers.</li>
    <li>sns.scatterplot: <strong>Scatter plots</strong> show the relationship between two continuous variables; if color-coded, we can also show the </li>relationship with a third categorical variable.
    <li>sns.regplot: including a <strong>regression line</strong> in the scatter plot makes it easier to see any linear relationship between two variables.</li>
    <li>sns.lmplot: this command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.</li>
    <li>sns.swarmplot: <strong>Categorical scatter plots</strong> show the relationship between a continous variable and a categorical variable. </li>
</ul>
<br/>


### Distribution 
<li>sns.histplot: <strong>Histograms</strong> show the distribution of a single numerical variable.</li>
<li>sns.kdeplot: <strong>KDE plots (or 2D KDE plots)</strong> show an estimated, smooth distribution of a single numerical variable (or two numerical variables).</li>
<li>sns.jointplot: this command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual </li>variable.
